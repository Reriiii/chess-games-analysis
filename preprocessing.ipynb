```python
import pandas as pd
import numpy as np
```

```python
DATE = "2025-01"
FILE = f"dataset/{DATE}.csv"
```

```python
df = pd.read_csv(FILE)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# xác định số cột chuẩn
columns = ["Event","Date","White_Name","Black_Name","WhiteElo","BlackElo",
           "WhiteRatingDiff","BlackRatingDiff","Result","Opening","TimeControl"]

# những dòng bị dồn hết vào Event
mask = df['Date'].isna()

# tách dữ liệu từ Event theo dấu phẩy
split_cols = df.loc[mask, 'Event'].str.split(',', expand=True)

# gán lại vào các cột tương ứng
df.loc[mask, columns] = split_cols.iloc[:, :len(columns)].values
```
